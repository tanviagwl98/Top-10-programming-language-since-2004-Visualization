```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("/kaggle/input/most-popular-programming-languages-since-2004/Most Popular Programming Languages from 2004 to 2020 V2.csv")
data.head()
```

```python
data.info
```

```python
data.describe()
```

```python
data["Date"] = pd.to_datetime(data["Date"])
data.set_index("Date", inplace = True)
data.head()
```

```python
plt.figure(figsize = (20,10))
sns_map = sns.heatmap(data, cmap="rainbow", yticklabels = False)
```

```python
data = data.loc[:, (data.mean() > 2.5)]
plt.figure(figsize = (10,10))

sns_map2 = sns.heatmap(data, center=0, yticklabels = False)
```

To get lineplot for top 10 programming language I considered **mean>2.5** above which number of programming languages were reducing.

```python
columns = data.columns.tolist()
```

```python
plt.figure(figsize = (30,20))

for lang in columns:
    ax = sns.lineplot(x = data.index, y = data[lang], label = lang, linewidth = 8)

ax.set_facecolor("black")
plt.title("Line Chart")
plt.ylabel("Popularity in %")
plt.xlabel("Time in date-time format")
legend = plt.legend(loc = 9, ncol = 10, fancybox = True, facecolor = '#ffffff', edgecolor = '#02bbd5', fontsize = 14)
plt.setp(legend.get_texts(), color='0')
plt.yticks()
plt.show()
```
